# Queries cypher

## Cria os nós do grafo

#### Carrega doenças
```
LOAD CSV WITH HEADERS FROM 'https://github.com/guilherme1905/Projeto-MC536/raw/master/jupyter/data/doencas.csv' AS line
CREATE (:Doenças {nome:line.Nome})
```
#### Carrega tipos de tratamento
```
LOAD CSV WITH HEADERS FROM 'https://github.com/guilherme1905/Projeto-MC536/raw/master/jupyter/data/tratamentos.csv' AS line
CREATE (:Tratamento {tipo:line.TRATAMENTO})
```
#### Cria aresta doença->tratamento
```
LOAD CSV WITH HEADERS FROM 'https://github.com/guilherme1905/Projeto-MC536/raw/master/jupyter/data/doencas.csv' AS line
MATCH (d:Doencas {nome:line.Nome})
MATCH (t:Tratamento {tipo:line.Tratamento})
CREATE (d)-[:tratado]->(t)
```
#### Carrega estados
```
LOAD CSV WITH HEADERS FROM 'https://github.com/guilherme1905/Projeto-MC536/raw/master/jupyter/data/uf.csv' AS line
CREATE (:UF {nome:line.Nome, regiao: line.Regiao, saneamento:line.Nota_saneamento, n_municipios:line.n_municipios})
```
#### Cria aresta estado->doença
```
LOAD CSV WITH HEADERS FROM 'https://github.com/guilherme1905/Projeto-MC536/raw/master/jupyter/data/doencas_por_estado.csv' AS line
MATCH (e:UF {nome:line.UF})
MATCH (d:Doencas {nome:line.Doenca})
CREATE (e)-[:infectado {n_casos:line.Casos}]->(d)
```

### Encontra os estados infectados por uma doença (Meningite)
```
MATCH p=(e:UF)-[i:infectado]->(d:Doencas)
WHERE d.nome = "Meningite"
RETURN p
```

![meningite](img/estados.png)

### Encontra as doenças que aparecem em um estado (Rondônia)
```
MATCH p=(e:UF)-[:infectado]-(:Doencas)
WHERE e.nome = "Rondonia"
RETURN p
```

![rondonia](img/doencas.png)

### Encontra os tratamentos necessários em um estado (Rondônia)
```
MATCH p=(e:UF)-[*]-(:Tratamento)
WHERE e.nome = "Rondonia"
RETURN p
```

![amazonas](img/tratamentos.png)

### Encontra os tipos de tratamentos mais necessários e as doenças que mais infectam
```
MATCH p=(:UF)-[:infectado]->(:Doencas)-[:tratado]->(:Tratamento)
RETURN p
```
